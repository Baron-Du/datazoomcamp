```python
df = pd.read_csv('green_tripdata_2019-10.csv')
```

```python
display(df)
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
df.to_sql(name='green_tripdata_trip_true', con=engine, index=False)
```

```python
query = """
SELECT 
    *
FROM green_tripdata_trip_true
limit 10
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT 
    count(case when trip_distance <= 1 THEN trip_distance else NULL end),
    count(case when trip_distance <= 3 and trip_distance > 1 THEN trip_distance else NULL end),
    count(case when trip_distance <= 7 and trip_distance > 3 THEN trip_distance else NULL end),
    count(case when trip_distance <= 10 and trip_distance > 7 THEN trip_distance else NULL end),
    count(case when trip_distance > 10 THEN trip_distance else NULL end)
FROM green_tripdata_trip_true
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT *
FROM green_tripdata_trip_true
order by trip_distance desc
limit 1
"""

pd.read_sql(query, con=engine)
```

```python
df = pd.read_csv('taxi_zone_lookup.csv')
```

```python
display(df)
```

```python
df.to_sql(name='taxi_zone_lookup', con=engine, index=False)
```

```python
query = """
with temp as (SELECT "PULocationID" as loc, sum(total_amount) as cnt
FROM green_tripdata_trip_true
where date(lpep_pickup_datetime) = '2019-10-18'
group by loc
order by cnt desc)

select a.loc, b."Zone" from temp a left join taxi_zone_lookup b on a.loc = b."LocationID" order by a.cnt desc
"""

pd.read_sql(query, con=engine)
```

```python
query = """
with temp as (SELECT "DOLocationID" as loc, sum(tip_amount) as tip, "PULocationID" as loc_1
FROM green_tripdata_trip_true
where "PULocationID" = 74
group by loc, loc_1
order by tip desc)

select a.loc, a.tip, a.loc_1, b."Zone" from temp a left join taxi_zone_lookup b on a.loc = b."LocationID" order by a.tip desc
"""

pd.read_sql(query, con=engine)
```
